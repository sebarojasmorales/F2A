```python
# Libraries
import numpy as np
import pandas as pd
```

```python
# Code to create a dataframe with all the Weather files 

df_weathers = pd.DataFrame()
n_scen=6 # Number of scenaries
n_weathers=100 # Number of Weather files for each scenario

for i in range(1,n_scen+1):
    for j in range(1,n_weathers+1):
        # Each file is read
        path = f'/home/sebar/Documents/SAESA/Final/escenario{i}/Weathers/Weather{j}.csv'
        df_local = pd.read_csv(path)
        
        # The scenario number is saved
        df_local['Scenario Group'] = i

        # The dataframe of the current file is merged with the dataframe containing all the information
        df_weathers = pd.concat([df_weathers,df_local]).reset_index(drop=True)
```

```python
# Boxplot with the maximum of the variable in each file
var = 'TM' # TM, RH, WS
df_weathers[['Scenario Group', 'Scenario', var]].groupby(['Scenario Group', 'Scenario']).max().boxplot(column=[var], by='Scenario Group')
```

```python
# Boxplot with all the data (and periods) from all the files
var = 'TM'
df_weathers[['Scenario Group', 'Scenario', var]].boxplot(column=[var], by='Scenario Group')
```
